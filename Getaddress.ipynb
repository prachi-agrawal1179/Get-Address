```python
from bs4 import BeautifulSoup


import urllib.request

response = urllib.request.urlopen('https://witsy.ai/')

html = response.read()

soup = BeautifulSoup(html,"html5lib")

text = soup.get_text(strip=True)

print (text)
```

```python
import spacy
from spacy.lang.en import English
import en_core_web_md
nlp = en_core_web_md.load()


my_doc = nlp(text.strip())

token_list = []
location_entities = set()
for i in my_doc.ents:
    entry = str(i.lemma_).lower()
    text = text.replace(str(i).lower(), "")
    if i.label_ in ["GPE", "GEO"]:
        location_entities.add(entry)
print(f"location entities - {location_entities}")
```
